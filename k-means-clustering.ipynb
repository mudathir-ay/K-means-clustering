```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler 
import os
import seaborn as sns
```

```python
iris_dataset = pd.read_csv("https://raw.githubusercontent.com/code-hack-dot-com/datasets/main/Iris.csv")
```

```python

# pd.set_option('display.max_rows', None)
iris_dataset
```

```python
iris_dataset.info()
```

```python
#Lets see how many classes they fall into
iris_dataset.value_counts(['Species'])
```

```python
#Lets say we're not sure of the datset we're dealing with and we decide to give k=2
km = KMeans(n_clusters=2)
x = iris_dataset.iloc[:, [1, 2, 3, 4]]
```

```python
x.head()
```

```python
sns.pairplot(iris_dataset,hue="Species",size=3)
```

```python
y_predicted = km.fit_predict(x)
y_predicted
```

```python
iris_dataset_copy = iris_dataset.copy()
iris_dataset_copy["2_k_prediction"] = y_predicted
```

```python
iris_dataset_copy.sample(20)
```

```python
sns.pairplot(iris_dataset_copy, hue = '2_k_prediction')
```

```python


cluster_one = iris_dataset_copy[iris_dataset_copy["2_k_prediction"]==0]
cluster_two = iris_dataset_copy[iris_dataset_copy["2_k_prediction"]==1]
plt.scatter(cluster_one["SepalLengthCm"], cluster_one["SepalWidthCm"])
plt.scatter(cluster_two["SepalLengthCm"], cluster_two["SepalWidthCm"])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color="black")
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.legend()
```

Why isnt our data well clustered? (one thing to consider is that our input data isnt normalized), another is that perhaps we have issues with the value of 'K' selected. Lets try the first assumption and see if its works

```python
scaler = MinMaxScaler()
scaler.fit(iris_dataset_copy.iloc[:, [1,2,3,4]])

iris_dataset_copy.iloc[:, [1,2,3,4]] = scaler.transform(iris_dataset_copy.iloc[:, [1,2,3,4]])
```

```python
iris_dataset_copy.head(20)
```

```python
#Lets try to predict
y_predicted_version2 = km.fit_predict(iris_dataset_copy.iloc[:,[1,2,3,4]])
y_predicted_version2
```

```python
iris_dataset_copy["2_k_prediction_with_nomalization"] = y_predicted_version2
iris_dataset_copy.sample(50)
#from 1d 58, we can see that normalizing gave some sort of difference
```

```python
cluster_three = iris_dataset_copy[iris_dataset_copy["2_k_prediction_with_nomalization"]==0]
cluster_four = iris_dataset_copy[iris_dataset_copy["2_k_prediction_with_nomalization"]==1]

plt.scatter(cluster_three["SepalLengthCm"], cluster_three["SepalWidthCm"])
plt.scatter(cluster_four["SepalLengthCm"], cluster_four["SepalWidthCm"])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'black')
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
```

There is a huge difference when normalization is applied. How about we use the elbow method to check the appropraite numer of K is right for this dataset?

```python
# wcss represents "within cluster sum of square", this basically computes the sum of square error for all the values of 'K'
wcss = []
for k in range(1,10):
    km = KMeans(n_clusters=k)
    km.fit(iris_dataset_copy.iloc[:,[1,2,3,4]])
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
#Lets plot so we can see where the elbow lies
plt.plot(range(1,10), wcss);
```

```python
#from above, we can see the elbow lies in in 3, so lets use 3 as k and see the difference.
km = KMeans(n_clusters= 3)
y_predicted_version3 = km.fit_predict(iris_dataset_copy.iloc[:,[1,2,3,4]])
iris_dataset_copy["3_k_prediction"] = y_predicted_version3
```

```python
iris_dataset_copy.sample(50)
```

```python
cluster_five = iris_dataset_copy[iris_dataset_copy["3_k_prediction"]==0]
cluster_six = iris_dataset_copy[iris_dataset_copy["3_k_prediction"]==1]
cluster_seven = iris_dataset_copy[iris_dataset_copy["3_k_prediction"]==2]

plt.scatter(cluster_five["SepalLengthCm"], cluster_five["SepalWidthCm"])
plt.scatter(cluster_six["SepalLengthCm"], cluster_six["SepalWidthCm"])
plt.scatter(cluster_seven["SepalLengthCm"], cluster_seven["SepalWidthCm"])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'black')
```
